## Information on storage power and capacity of the storage provider

```python
import requests
import datetime
import json

# Defines the fixed part of the url
base_url = "https://api.spacescope.io/v2/storage_provider/power?state_date="

# Defines the start date and end date
start_date = datetime.date(2021, 1, 1)
end_date = datetime.date(2023, 2, 20)

# Defines the payload and headers
payload={}
headers = {
  'authorization': 'Bearer ghp_ypbWCaRHdYRXeuVjftzHkjkLZmVrkFKindOQ'
}

# Creates an empty list to store the results
results = []

# Creates a for loop to iterate through the dates between the start date and end date
for n in range((end_date - start_date).days + 1):
  # Generates the intermediate date
  date = start_date + datetime.timedelta(n)

  # Formats the date as a string in the format "yyyy-mm-dd"
  date_str = date.strftime("%Y-%m-%d")

  # Concatenates the string with the fixed part of the url
  url = base_url + date_str

  # Makes the request and get the Response object
  response = requests.request("GET", url, headers=headers, data=payload)

  # Converts the text of the Response object into a dictionary
  data = response.json()

  # Adds the dictionary to the list of results
  results.append(data)

  # Prints a progress message every 10 iterations
  if n % 10 == 0:
    print(f"Progress: {n}/{(end_date - start_date).days + 1}")

# Converts the list of results into a JSON
results_json = json.dumps(results)

# Prints the JSON with all the results
print(results_json)
```

```python
import pandas as pd
import json

# Converte o results_json em uma lista de dicionários
results_list = json.loads(results_json)

# Cria uma lista vazia para armazenar os dicionários formatados
formatted_list = []

# Itera sobre cada dicionário na lista de resultados
for result in results_list:
  # Itera sobre cada dicionário na lista result['data']
  for data in result['data']:
    # Cria um dicionário vazio para armazenar os valores formatados
    formatted_dict = {}

    # Adiciona os valores das 7 colunas desejadas ao dicionário formatado
    formatted_dict['request_id'] = result['request_id']
    formatted_dict['code'] = result['code']
    formatted_dict['message'] = result['message']
    formatted_dict['stat_date'] = data['stat_date']
    formatted_dict['miner_id'] = data['miner_id']
    formatted_dict['raw_byte_power'] = data['raw_byte_power']
    formatted_dict['quality_adj_power'] = data['quality_adj_power']

    # Adiciona o dicionário formatado à lista de dicionários formatados
    formatted_list.append(formatted_dict)

# Cria um dataframe pandas a partir da lista de dicionários formatados
df = pd.DataFrame(formatted_list)

# Imprime os primeiros 5 registros
df.head()
```

```python
# Baixa o dataframe como um arquivo csv chamado 'results.csv'
df.to_csv('results.csv', sep=',', index=False)
```
